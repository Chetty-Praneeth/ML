```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('k_means.csv')
df
```

```python
plt.scatter(df['feature1 '],df['feature2'])
```

```python
km = KMeans(n_clusters=3, random_state=0)
km
```

```python
y_pred = km.fit_predict(df[['feature1 ','feature2']])
```

```python
y_pred
```

```python
df['cluster'] = y_pred
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1['feature1 '],df1.feature2, color = 'green')
plt.scatter(df2['feature1 '],df2.feature2, color = 'red')
plt.scatter(df3['feature1 '],df3.feature2, color = 'blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = "yellow",marker = '+')
plt.legend()
```
